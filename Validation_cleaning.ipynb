```python
import pandas as pd
```

```python
df = pd.read_excel('EVALUATION.xlsx')
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.drop('BagsCount', axis=1, inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df.shape
```

```python
df['FLownPassengers'] = pd.to_numeric(df['FLownPassengers'], errors='coerce')
mean_FLownPassengers = df['FLownPassengers'].mean()
df['FLownPassengers'].fillna(mean_FLownPassengers, inplace=True)
```

```python
df['FLownPassengers'].unique()
```

```python
df['FlightBagsWeight'] = pd.to_numeric(df['FlightBagsWeight'], errors='coerce')
mean_FlightBagsWeight = df['FlightBagsWeight'].mean()
df['FlightBagsWeight'].fillna(mean_FlightBagsWeight, inplace=True)
```

```python
df['FlightBagsWeight'].describe()
```

```python
df.dropna(dropna=True, inplace=True)
```

```python
df['DepartureDate'] = pd.to_datetime(df['DepartureDate'], format="%d.%m.%Y")
df['DayOfWeek'] = df['DepartureDate'].dt.day_name()
df['Weekend'] = df['DayOfWeek'].isin(['Friday','Saturday', 'Sunday'])
df['WeekDay'] = df['DayOfWeek'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday'])
df.drop(['DepartureDate', 'DayOfWeek'], axis=1, inplace=True)
```

```python
df['Weekend'] = df['Weekend'].apply(lambda x: 1 if x in [True, 'True'] else 0)
df['WeekDay'] = df['WeekDay'].apply(lambda x: 1 if x in [True, 'True'] else 0)
```

```python
df
```

```python
df.drop('DepartureAirport', axis=1, inplace=True)
```

```python
df.drop('ArrivalAirport', axis=1, inplace=True)
```

```python
eva_df=df
```

```python
df.to_csv('result.csv')
```
